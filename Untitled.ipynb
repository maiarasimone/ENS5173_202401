```python
# criando uma lista de python
list = [0,'eu amo polAtm', 2.5]
print(list)
#Nesta parte fiz a diação entre duas strings, sendo que list[1] é o segundo indíce do vetor list
print('variável: '+ list[1])

#Adição do primeiro indice da variável list (list[0]) - Lembrar que a indexação no python inicia em 0
print(list[0]+1)
```

```python
# Criando uma matriz numpy

import numpy as np
#Criando uma matriz de zeros utilizando np.zeros
a = np.zeros((5,5))
print(a)
# pirmeiro linha, depois coluna - linha 0, coluna 0 = item 0
print('indexação numpy')
print(a[0,0])
```

```python
# Criando números aleatórios
b = np.random.rand(5,5)
print(b)
print('Dimensão da matriz')
#dimensão da matriz
print(b.shape)

print(type(b))
```

```python
#Criando matriz de valores aleatórios entre uma faixa estabelecida
vmin=1 #valor mínimo
vmax=100 #valor máximo
#Matriz
c= vmin+(vmax-vmin)*b
print(c)
```

```python
#Plotar matriz
import matplotlib.pyplot as plt


plt.plot(c.transpose())
```

```python
#Plotando apenas a primeira linha
x= np.array(range(0,np.size(c[0,:])))
print(x)
plt.plot(c[0,:])
plt.title('Gráfico com Valores Inteiros no Eixo x')
plt.xticks(x) #Definindo os valores do eixo x
```

```python
#Criando um gráfico de barras e exportando
#ax são os eixos

fig, ax = plt.subplots(3,1)
ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

#exportar a figura
import os
rootFolder = os.getcwd()
print(rootFolder)
figFolder = rootFolder.replace('\\','/')+'/figures/'
print(figFolder)
fig.savefig(figFolder+'Figure01.png')
```
