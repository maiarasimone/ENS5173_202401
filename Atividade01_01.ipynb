```python
# Utilizando Pandas para abrir a planilha
import pandas as pd
import os


# Coletando o caminho da pasta que estou
cwd = os.getcwd()

# Pasta dos fatores de emiss√£o
path = cwd+'\\tabelas\\FE'
print(path)


# Indicando nome do arquivo
chapterNumer = 10
subChapter = 2
file = 'chp_'+str(chapterNumer)+'_'+str(subChapter)+'.csv'
print(file)

fe =pd.read_csv(path + '\\' +file)
print(fe)
```

```python
# Plotando os fatores de emiss√£o
import matplotlib.pyplot as plt
#298271
atv = 817 #taxa de atividade
emis = fe*atv


#somando dataframe
print(emis.sum())

#criando a figura com matplotlib
#criando subplots com numero de linhas igual ao numero de poluentes 
fig,ax = plt.subplots(len(fe.columns)-1,1,figsize=(12, 15))

#loop para reencher o subplot com os graficos de barra : Para cada itera√ß√£o do loop, um gr√°fico de barra √© criado usando o m√©todo .
#bar do objeto Axes (presumivelmente ax √© uma lista de objetos Axes). O eixo x √© definido como os valores da coluna 'Source', 
#e o eixo y √© definido como os valores da coluna atual (fe.iloc[:, ii]). A cor das barras √© definida manualmente.
for ii in range(1,len(fe.columns)):
    print(ii)
    ax[ii-1].bar(fe['Source'],fe.iloc[:,ii], color=['pink', 'violet','cyan','blue','yellow', 'green', 'purple', 'gold'])
    ax[ii-1].set_xlabel('Processo', fontsize = 10)
    ax[ii-1].set_ylabel(fe.columns[ii]+'\n (kg/Mg)', fontsize = 10) 

#salvando a figura na pasta figures
fig.savefig(cwd+'\\figures\\barplot_fatorEmiss.png')















```

```python
import numpy as np


# Definindo a porcentagem para calcular a atividade m√°xima e m√≠nima (80% neste caso)
porcentagem = 0.8

# Calculando a atividade m√°xima e m√≠nima
atvMax = atv* (1 + porcentagem)
atvMin = atv * (1 - porcentagem)

# Mostrando os resultados
print("Atividade m√°xima:", atvMax)
print("Atividade m√≠nima:", atvMin)

# Criando vetor de valores aleat√≥rios
n = 1000
atv = (atvMax-atvMin)*np.random.rand(n) + atvMin

# Criando uma figura com os valores aleat√≥rios
fig, ax= plt.subplots()
ax.plot(atv)


```

```python
# Criar um vetor com as datas
from datetime import datetime, timedelta

# Definindo a data inicial e final
data_inicial = datetime(2023, 1, 1)
data_final = datetime(2023, 12, 31)

# Definindo o passo (intervalo entre as datas)
passo = timedelta(days=1)

# Criando o vetor de datas
vetor_datas = []
data_atual = data_inicial
while data_atual <= data_final:
    vetor_datas.append(data_atual)
    data_atual += passo

# Exibindo o vetor de datas
for data in vetor_datas:
    print(data)
```

```python
# Criando vetor de datas
import pandas as pd

# Criar um intervalo de datas
datas = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

# Criar um DataFrame com as datas
df = pd.DataFrame({'Data': datas})

# Extrair os meses das datas e adicion√°-los como uma nova coluna no DataFrame
df['Mes'] = df['Data'].dt.month
df['Dia do Ano'] = df['Data'].dt.dayofyear  #criar um array de 1 a 365 e colocar igual a df['Dia']
# Mostrar o DataFrame resultante
print(df)

# Mostrar o DataFrame resultante
print(df)
```

```python
# Ajustar meus n√∫meros aleat√≥rios com base no comprimento do vetor de datas

# Criando dataframe com datas e valores de atividade aleat√≥ria
import pandas as pd
# Extraindo o n√∫mero de linhas/n√∫mero de datas
n = df.shape[0]
# Criando atividade aleat√≥ria
atv = (atvMax - atvMin)*np.random.rand(n) + atvMin
df['atividade'] = atv
#print(atv)
print(df)

df.to_csv("C:\\Users\\sergi\\OneDrive\\Documentos\\GitHub\\ENS5173_202401\\tabelas\\ativ.csv")
```
