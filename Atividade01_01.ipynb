```python
import numpy as np


# Definindo a porcentagem para calcular a atividade m√°xima e m√≠nima (80% neste caso)
porcentagem = 0.8

# Calculando a atividade m√°xima e m√≠nima
atvMax = atv* (1 + porcentagem)
atvMin = atv * (1 - porcentagem)

# Mostrando os resultados
print("Atividade m√°xima:", atvMax)
print("Atividade m√≠nima:", atvMin)

# Criando vetor de valores aleat√≥rios
n = 1000
atv = (atvMax-atvMin)*np.random.rand(n) + atvMin

# Criando uma figura com os valores aleat√≥rios
fig, ax= plt.subplots()
ax.plot(atv)


```

```python
# Criar um vetor com as datas
from datetime import datetime, timedelta

# Definindo a data inicial e final
data_inicial = datetime(2023, 1, 1)
data_final = datetime(2023, 12, 31)

# Definindo o passo (intervalo entre as datas)
passo = timedelta(days=1)

# Criando o vetor de datas
vetor_datas = []
data_atual = data_inicial
while data_atual <= data_final:
    vetor_datas.append(data_atual)
    data_atual += passo

# Exibindo o vetor de datas
for data in vetor_datas:
    print(data)
```

```python
# Criando vetor de datas
import pandas as pd

# Criar um intervalo de datas
datas = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

# Criar um DataFrame com as datas
df = pd.DataFrame({'Data': datas})

# Extrair os meses das datas e adicion√°-los como uma nova coluna no DataFrame
df['Mes'] = df['Data'].dt.month
df['Dia do Ano'] = df['Data'].dt.dayofyear  #criar um array de 1 a 365 e colocar igual a df['Dia']
# Mostrar o DataFrame resultante
print(df)

# Mostrar o DataFrame resultante
print(df)
```

```python
# Ajustar meus n√∫meros aleat√≥rios com base no comprimento do vetor de datas

# Criando dataframe com datas e valores de atividade aleat√≥ria
import pandas as pd
# Extraindo o n√∫mero de linhas/n√∫mero de datas
n = df.shape[0]
# Criando atividade aleat√≥ria
atv = (atvMax - atvMin)*np.random.rand(n) + atvMin
df['atividade'] = atv
#print(atv)
print(df)

df.to_csv("C:\\Users\\sergi\\OneDrive\\Documentos\\GitHub\\ENS5173_202401\\tabelas\\ativ.csv")
```

```python
# Criando componente sazonal

sazonalidade = np.cos((2*np.pi)*df['Dia do Ano']/365)+1 #taxa de papel produzido diariamente
df['sazonalidade'] = sazonalidade
plt.plot(sazonalidade)
print(df)
df.to_csv(r"C:\Users\sergi\OneDrive\Documentos\GitHub\ENS5173_202401\tabelas\ativ.csv")
```

```python
import matplotlib.pyplot as plt

# Calcula os valores aleat√≥rios com sazonalidade e os atribui √† coluna 'aleatorioSazonal' do DataFrame `df`, resultado √© multiplicado por 0.4948 - n√∫mero escolhido para chegar ao valor pr√≥ximo da m√©dia anual 
df['aleatorioSazonal'] = (df['atividade'] +  df['sazonalidade']*df['atividade'].mean())*0.505

# Plota o gr√°fico
plt.figure(figsize=(8, 6))
plt.plot(df['Dia do Ano'], df['aleatorioSazonal'])

# Definindo o t√≠tulo e os r√≥tulos dos eixos
plt.xlabel('Dias no ano')
plt.ylabel('Atividade (kg)')

# Exibindo o gr√°fico
plt.savefig(r"C:\Users\sergi\OneDrive\Documentos\GitHub\ENS5173_202401\figures\atividade.png")
plt.show()

# Calculando a m√©dia dos valores em 'aleatorioSazonal' e multiplicando por 365 para obter uma m√©dia anual
media_anual_kg = df['aleatorioSazonal'].mean() * 365
print("M√©dia anual em kg:", media_anual_kg)

# Salvando o DataFrame `df` como um arquivo CSV
df.to_csv(r"C:\Users\sergi\OneDrive\Documentos\GitHub\ENS5173_202401\tabelas\ativ.csv")

   
```

```python


import matplotlib.pyplot as plt

# Taxas de emiss√£o dos poluentes para cada processo
emissoes_processos = {
    'recovery boiler': {'MP': 90.0, 'SO2': 3.50, 'CO': 5.50},
    'smelt tank': {'MP': 3.5, 'SO2': 0.10, 'CO': 0.00},
    'lime kiln': {'MP': 28.0, 'SO2': 0.15, 'CO': 0.05}
}

# Loop para cada processo
for processo, emissoes in emissoes_processos.items():
    # Cria uma nova figura para o processo atual
    fig, axs = plt.subplots(3)
    fig.suptitle(f'Emiss√µes de Poluentes para o Processo {processo}')
    # Loop para cada poluente
    for i, (poluente, taxa_emissao) in enumerate(emissoes.items()):
        # Multiplica a taxa de emiss√£o pelo fator aleat√≥rio sazonal para o processo e poluente espec√≠fico
        emissao_processo_poluente = taxa_emissao * df['aleatorioSazonal']
        # Plota o gr√°fico para o poluente atual no subplot correspondente
        axs[i].plot(df['Dia do Ano'], emissao_processo_poluente)
        # Adiciona r√≥tulos aos eixos
        axs[i].set_xlabel('Dia do Ano')
        axs[i].set_ylabel(f'{poluente}(kg)')

# Ajusta o layout dos subplots para evitar sobreposi√ß√£o
plt.tight_layout()
# Exibe os gr√°ficos
plt.show()

```

```python


import numpy as np

# Supondo que voc√™ tenha um fator aleat√≥rio sazonal
# Vou usar valores aleat√≥rios como exemplo
fator_aleatorio_sazonal = df['aleatorioSazonal_g_s']

# Taxas de emiss√£o dos poluentes para cada processo (em g)
emissoes_processos = {
    'recovery boiler': {'MP': 90.0 * 1000, 'SO2': 3.50 * 1000, 'CO': 5.50 * 1000},
    'smelt tank': {'MP': 3.5 * 1000, 'SO2': 0.10 * 1000, 'CO': 0.00},
    'lime kiln': {'MP': 28.0 * 1000, 'SO2': 0.15 * 1000, 'CO': 0.05 * 1000}
}

# Dicion√°rio para armazenar m√©dia, m√≠nimo e m√°ximo de emiss√£o para cada processo e poluente
estatisticas_emissoes = {}

# Loop para cada processo
for processo, emissoes in emissoes_processos.items():
    # Dicion√°rio para armazenar estat√≠sticas de emiss√£o para o processo atual
    estatisticas_emissoes[processo] = {}
    # Loop para cada poluente
    for poluente, taxa_emissao in emissoes.items():
        # Multiplica as taxas de emiss√£o pelo fator aleat√≥rio sazonal para o processo e poluente espec√≠fico
        emissao_processo_poluente = taxa_emissao * df['aleatorioSazonal_g_s']
        # Calcula m√©dia, m√≠nimo e m√°ximo de emiss√£o para o processo e poluente atual
        media_emissao = np.mean(emissao_processo_poluente)
        minima_emissao = np.min(emissao_processo_poluente)
        maxima_emissao = np.max(emissao_processo_poluente)
        # Armazena as estat√≠sticas de emiss√£o para o processo e poluente atual
        estatisticas_emissoes[processo][poluente] = {'m√©dia': media_emissao, 'm√≠nima': minima_emissao, 'm√°xima': maxima_emissao}

# Imprime as estat√≠sticas de emiss√£o para cada processo e poluente
for processo, estatisticas_poluente in estatisticas_emissoes.items():
    print(f"\nEmiss√£o para o processo '{processo}':")
    for poluente, estatisticas in estatisticas_poluente.items():
        print(f"Poluente: {poluente}")
        print("M√©dia: {:.4f} g/s".format(estatisticas['m√©dia']))
        print("M√≠nima: {:.4f} g/s".format(estatisticas['m√≠nima']))
        print("M√°xima: {:.4f} g/s".format(estatisticas['m√°xima']))


```
