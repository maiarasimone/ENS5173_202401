## Gerar números aleatórios que representem a atividade industrial

A empresa escolhida foi a Cerâmica Santa Catarina, referente à produção de cerâmicas

```python
import pandas as pd
import numpy as np
import math as mt
import matplotlib.pyplot as plt
import random




```

```python
periodo = pd.date_range(start='1/1/2024', end='31/12/2024', freq='ME')
```

```python
# Simulando a produção mensal
producao_mensal = [random.randint(1000, 5000) for _ in range(len(periodo))]
```

```python
# Criando o DataFrame
df = pd.DataFrame({
    'Data': periodo,
    'Produção Mensal': producao_mensal
})
```

```python
# Definindo 'Data' como índice
df.set_index('Data', inplace=True)
```

```python
print(df)
```

```python
import matplotlib.pyplot as plt

# Criando o gráfico
plt.figure(figsize=(10,6))
plt.plot(df.index, df['Produção Mensal'], marker='o', linestyle='-')
plt.title('Produção Mensal de Cerâmica')
plt.xlabel('Data')
plt.ylabel('Produção Mensal')
plt.grid(True)
plt.show()
```
